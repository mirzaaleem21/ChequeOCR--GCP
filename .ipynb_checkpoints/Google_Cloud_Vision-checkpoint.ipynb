```python
# Authentication to Google API
import os
import math
from collections import Counter
from google.cloud import vision
import re

os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = '/Users/mirza/Documents/Spring 2025/OCr/vision_key.json'

WORD = re.compile(r"\w+")
```

```python
def detect_text(path):
    """Detects text in the file."""

    client = vision.ImageAnnotatorClient()

    with open(path, "rb") as image_file:
        content = image_file.read()

    image = vision.Image(content=content)

    # for non-dense text 
    # response = client.text_detection(image=image)
    # for dense text
    response = client.document_text_detection(image=image)
    texts = response.text_annotations
    ocr_text = []

    for text in texts:
        ocr_text.append(f"\r\n{text.description}")

    if response.error.message:
        raise Exception(
            "{}\nFor more info on error messages, check: "
            "https://cloud.google.com/apis/design/errors".format(response.error.message)
        )
    return ocr_text
```

```python
def extract_check_details(text_with_boxes):  # Now takes text with boxes
    details = {}
       

    # *** Correct way to handle the list of strings ***
    if isinstance(text_with_boxes, list):  # Check if it's a list
        text_lines = text_with_boxes  # Use the list directly
    elif isinstance(text_with_boxes, str): # Check if its a string
        text_lines = text_with_boxes.splitlines() # split the string to lines
    else:
        return {}  # Or handle the error as you see fit
    text_lines = text_with_boxes.splitlines()

    # 1. Recipient (Pay To)
    for line in text_lines:
        if "PAY TO THE ORDER OF" in line or "Pay To" in line or "Pay to" in line:
            recipient_match = re.search(r"(PAY TO THE ORDER OF|Pay To|Pay to)(.*?)(\n|$)", line)
            if recipient_match:
                details['recipient'] = recipient_match.group(2).strip()
                break  # Stop searching once found

    # 2. Amount (Numerical)
    for line in text_lines:
        amount_match = re.search(r"\$\*\*?([\d,]+(\.\d{2})?)", line)  # Prioritize with **
        if amount_match:
            details['amount'] = amount_match.group(1).replace(",", "")
            break
        else:
            amount_match = re.search(r"\$([\d,]+(\.\d{2})?)", line)  # Then try without **
            if amount_match:
                details['amount'] = amount_match.group(1).replace(",", "")
                break

    # 3. Date
    for line in text_lines:
        date_match = re.search(r"(DATE|Date|date)(.*?)(\n|$)", line)
        if date_match:
            details['date'] = date_match.group(2).strip()
            break

    # 4. Memo
    for line in text_lines:
        memo_match = re.search(r"(MEMO|Memo|memo)(.*?)(\n|$)", line, re.IGNORECASE)
        if memo_match:
            details['memo'] = memo_match.group(2).strip()
            break

    # 5. Issuer (From Address) -  Improved using line proximity to other fields
    issuer_lines = []
    potential_issuer = []
    found_pay_to = False
    found_date = False

    for line in text_lines:
        if "PAY TO THE ORDER OF" in line or "Pay To" in line or "Pay to" in line:
            found_pay_to = True
        if "DATE" in line or "Date" in line or "date" in line:
            found_date = True
        
        if not found_pay_to and not found_date: # if pay to and date are not found, it is most likely issuer
            potential_issuer.append(line)

    details['issuer'] = "\n".join(potential_issuer).strip()

    return details
```

```python
def get_cosine(vec1, vec2):
    intersection = set(vec1.keys()) & set(vec2.keys())
    numerator = sum([vec1[x] * vec2[x] for x in intersection])

    sum1 = sum([vec1[x] ** 2 for x in list(vec1.keys())])
    sum2 = sum([vec2[x] ** 2 for x in list(vec2.keys())])
    denominator = math.sqrt(sum1) * math.sqrt(sum2)

    if not denominator:
        return 0.0
    else:
        return float(numerator) / denominator


def text_to_vector(text):
    words = WORD.findall(text)
    return Counter(words)
```

```python
text1 = "This is a foo bar sentence ."
text2 = "This sentence is similar to a foo bar sentence ."

text1 = "AUSTIN FORTE Notary Public - State of Colorado "
text2 = "Austin forth Notary Public - State of Colorado "

vector1 = text_to_vector(text1)
vector2 = text_to_vector(text2)

cosine = get_cosine(vector1, vector2)

print("Cosine:", cosine)
```
